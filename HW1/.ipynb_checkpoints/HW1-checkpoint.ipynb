## Hybrid Image

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (5, 5)
```

```python
def read_image(file_name, grayscale=True):
    # OpenCV uses BGR as its default colour order for images, matplotlib uses RGB.
    # flag ref: https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_image_display/py_image_display.html
    if not grayscale:
        img = cv2.imread(file_name, cv2.IMREAD_UNCHANGED)  # This is BGR
        b,g,r = cv2.split(img)
        rgb_img = cv2.merge([r,g,b])  # This is RGB
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        return img, rgb_img, gray_img
    else:
        img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)  # This is BGR
        return img
```

```python
def get_magnitude(img):
    dft = np.fft.fft2(img)
    dft_shift = np.fft.fftshift(dft)  # shift the zero frequency to center
    magnitude = 20*np.log(np.abs(dft_shift))  # why we have to perform log transformation?
    return magnitude
```

```python
def bgr2rgb(img):
    b,g,r = cv2.split(img)
    rgb_img = cv2.merge([r,g,b])  # This is RGB
    return rgb_img
```

```python
def plot_mag(laplacian, gaussian):
    fig, ax = plt.subplots(1, 3, figsize=[8, 6])
    ax[0].imshow(get_magnitude(laplacian), cmap='gray')
    ax[1].imshow(get_magnitude(gaussian), cmap='gray')
    ax[2].imshow(get_magnitude(normalize(laplacian + gaussian).astype(int)), cmap='gray')
    plt.tight_layout()
```

```python
def plot_img(laplacian, gaussian, cmap=None):
    fig, ax = plt.subplots(1, 3, figsize=[8, 6])
    ax[0].imshow(laplacian.astype(int), cmap=cmap)
    ax[1].imshow(gaussian, cmap=cmap)
    ax[2].imshow(normalize(laplacian + gaussian).astype(int), cmap=cmap)
    plt.tight_layout()
```

```python
def normalize(x):
    x = (x - x.min()) / (x.max() - x.min())
    x = (x * (255 - 0)) + 0
    return x
```

```python
def merge_rgb_img(img1_filename, img2_filename, scale=7):
    img1, rgb_img1, gray_img1 = read_image(img1_filename, grayscale=False)
    img2, rgb_img2, gray_img2 = read_image(img2_filename, grayscale=False)
    
    rgb_img2 = cv2.resize(rgb_img2,(img1.shape[1], img1.shape[0]), interpolation = cv2.INTER_CUBIC)
    gray_img2 = cv2.resize(gray_img2,(img1.shape[1], img1.shape[0]), interpolation = cv2.INTER_CUBIC)
    r, g, b = cv2.split(rgb_img1)
    laplacian_b = cv2.Laplacian(b, cv2.CV_64F, ksize=scale)
    laplacian_g = cv2.Laplacian(g, cv2.CV_64F, ksize=scale)
    laplacian_r = cv2.Laplacian(r, cv2.CV_64F, ksize=scale)
    gaussian = cv2.GaussianBlur(rgb_img2,(scale+2, scale+2),0)

    # normalize to [0, 255]
    laplacian = np.array([laplacian_r, laplacian_g, laplacian_b])
    laplacian = normalize(laplacian).astype(int)
    laplacian = cv2.merge(laplacian)
    plot_img(laplacian, gaussian)
```

```python
def merge_gray_img(img1_filename, img2_filename, scale=7):
    img1 = read_image(img1_filename)
    img2 = read_image(img2_filename)
    img2 = cv2.resize(img2,(img1.shape[1], img1.shape[0]), interpolation = cv2.INTER_CUBIC)

    laplacian = cv2.Laplacian(img1, cv2.CV_64F, ksize=scale)
    gaussian = cv2.GaussianBlur(img2,(scale+2, scale+2),0)
    
    laplacian = normalize(laplacian).astype(int)
    plot_mag(laplacian, gaussian)
    plot_img(laplacian, gaussian, cmap='gray')
```

```python
merge_gray_img('./image/makeup_after.jpg', './image/makeup_before.jpg')
merge_rgb_img('./image/makeup_after.jpg', './image/makeup_before.jpg')
```

```python
merge_gray_img('./image/Afghan_girl_after.jpg', './image/Afghan_girl_before.jpg')
merge_rgb_img('./image/Afghan_girl_after.jpg', './image/Afghan_girl_before.jpg')
```
